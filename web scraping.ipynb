# IMDB

# Mini Project 01 - IMDB web scraping

```r
library(tidyverse)
library(rvest) #scrape data from internet
```

```r
url <- "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating%2Cdesc"
```

```r
print(url)
```

```r
#read html
imdb <- read_html(url)
```

```r
imdb
```

```r
# movie title
title <- imdb %>%
    html_nodes("h3.lister-item-header") %>%
    html_text2()
```

```r
title[1:10]
```

```r
#rating
rating <- imdb %>%
    html_nodes("div.ratings-imdb-rating") %>%
    html_text2() %>%
    as.numeric()
```

```r
rating[1:10]
```

```r
#number of votes
num_votes <- imdb %>%
    html_nodes("p.sort-num_votes-visible") %>%
    html_text2()
```

```r
num_votes[1:10]
```

```r
#build a dataset
df <- data.frame(
    title = title,
    rating = rating,
    num_votes = num_votes
)

head(df)
```

# **Mini Project 02 - Specphone Phone Database**

```r
library(tidyverse)
library(rvest) #scrape data from internet
```

```r
url = "https://specphone.com/Samsung-Galaxy-A04.html"
```

```r
att <- url %>%
    read_html() %>%
    html_nodes("div.topic") %>%
    html_text2()

value <- url %>%
    read_html() %>%
    html_nodes("div.detail") %>%
    html_text2()
```

```r
data.frame(att = att, value = value)
```

```r
#all samsung Smartphones
samsung_url = "https://specphone.com/brand/Samsung" %>% read_html()
```

```r
#link to all samsung smartphone
links <- samsung_url %>%
    html_nodes("li.mobile-brand-item a") %>%
    html_attr("href")
```

```r
full_links <- paste0("https://specphone.com",links)
```

```r
full_links[1:10]
```

```r
for(link in full_links){
    print(link)
}
```

```r
result <- data.frame()
for(link in full_links){
    ss_topic <- link %>%
        read_html() %>%
        html_nodes("div.topic") %>%
        html_text2()

    ss_detail <- link %>%
        read_html() %>%
        html_nodes("div.detail") %>%
        html_text2()
    
    tmp <- data.frame(attribute = ss_topic,value = ss_detail)
    
    result <- bind_rows(result, tmp)
}
print(result)
```

```r
#write_csv
write_csv(result,"result_ss_phone.csv")
```
